1) 단지명 중  '해밀마을 2단지'인 row 삭제

2) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경

```python
df['new_시군구'] = df['시군구']
df.loc[df['new_시군구'] == '세종특별자치시 고운동', 'new_시군구'] = '세종특별자치시 해밀동'
```

```python
df
```

(3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 

```python
df.loc[df['전월세구분']=='전세', 'Class'] = True
df.loc[df['전월세구분']=='월세', 'Class'] = False
```

```python
df['Class'] = False
df.loc[df['전월세구분'] == '전세', 'Class'] = True
```

5) -10 ~ 0층 : 지하 , 1 ~ 5층 : 저층 6 ~ 15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 

```python
cond1 = df['층'] <= 0
cond2 = (df['층'] >= 1) & (df['층'] <= 5)
cond3 = (df['층'] >= 6) & (df['층'] <= 15)
cond4 = (df['층'] >= 16)

# condition_lst = [(df['층'] <= 0),
#                  (df['층'] >= 1) & (df['층'] <= 5),
#                  (df['층'] >= 6) & (df['층'] <= 15),
#                  (df['층'] >= 16)]

# df.loc[cond1, '층구분'] = '지하'
# df.loc[cond2, '층구분'] = '저층'
# df.loc[cond3, '층구분'] = '중층'
# df.loc[cond4, '층구분'] = '고층'

con_lst = [cond1, cond2, cond3, cond4]
층_lst = ['지하', '저층', '중층', '고층']

for i in range(4):
    df.loc[con_lst[i], '층구분'] = 층_lst[i]
```

```python
df
```

```python
import numpy as np
df['새 층 구분'] = np.select(con_lst, 층_lst, default = 'Not_Specified')
df
```
