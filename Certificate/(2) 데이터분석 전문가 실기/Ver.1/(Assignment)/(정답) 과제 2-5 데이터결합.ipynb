## sofifa_id를 key값으로  위의 두 테이블을 innerjoin 하세요 

```python
df19 = fifa_df19.loc[:, ['sofifa_id', 'short_name', 'overall']]
df20 = fifa_df20.loc[:, ['sofifa_id', 'overall']]
inner_merge = pd.merge(df19, df20, how='inner', on='sofifa_id')
inner_merge.columns = ['sofifa_id', 'short_name', 'overall_19', 'overall_20']
inner_merge
```

## sofifa_id를 key값으로  위의 두 테이블을 outerjoin 하세요 

```python
outer_merge = pd.merge(df19, df20, how='outer', on='sofifa_id')
outer_merge.columns = ['sofifa_id', 'short_name', 'overall_19', 'overall_20']
outer_merge
```

## outerjoin을 한 테이블 중 inner join을 한 테이블과 중복되는 행을 삭제하세요 

방법1

```python
outer_merge[~outer_merge['sofifa_id'].isin(inner_merge['sofifa_id'])]
```

방법2

```python
new1 = pd.concat([inner_merge, outer_merge], ignore_index = True)
new1
```

```python
new1.drop_duplicates(ignore_index = True)
```

방법3

```python
remove_idx = inner_merge['sofifa_id']
remove_idx
```

```python
outer_merge.set_index('sofifa_id', inplace = True)
outer_merge
```

```python
outer_merge.drop(remove_idx, inplace = True)
outer_merge
```
