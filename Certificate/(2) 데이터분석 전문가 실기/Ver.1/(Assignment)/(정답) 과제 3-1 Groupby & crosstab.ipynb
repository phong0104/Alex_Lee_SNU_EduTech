## 전세의 보증금 평균 값을 구하세요 groupby 사용 


```python
import pandas as pd
df = pd.read_csv('../data/7.세종시_아파트(전월세)_실거래가.csv')
df
```

```python
df.groupby(['전월세구분'])[['보증금(만원)']].mean()
```

```python
df.groupby(['전월세구분']).mean()[['보증금(만원)']].loc[['전세']]
```

```python
df.groupby(['전월세구분'])['보증금(만원)'].mean().values
```

```python
print(df.groupby(['전월세구분'])['보증금(만원)'].mean().values[1])
```

## 계약년도별, 시군구별 평균 보증금 가격를 구하세요  



```python
df.head()
```

```python
# del df['Unnamed: 0']
df.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
df.head()
```

```python
df['계약년월'].astype(str).str[:4]
```

```python
df['계약년'] = df['계약년월'].astype(str).str.slice(start=0, stop=4)
```

```python
df['계약년도'] = df['계약년월'].astype(str).str[:4]
```

```python
df.head()
```

```python
df.groupby(['계약년도', '시군구'])[['보증금(만원)']].mean()
```

```python
df.loc[df['전월세구분'] == '전세'].groupby(['계약년도']).count()
```
