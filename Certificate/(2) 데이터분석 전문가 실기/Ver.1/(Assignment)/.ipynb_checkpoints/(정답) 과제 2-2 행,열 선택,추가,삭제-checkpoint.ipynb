1) 세종시 데이터에서 컬럼 확인 후, unnamed 컬럼 제거

```python
import pandas as pd
df = pd.read_csv('../data/7.세종시_아파트(전월세)_실거래가.csv')
df.head()
```

2) 전월세구분 컬럼에서 unique값 확인 

```python
df['전월세구분'].unique()
```

```python
df[['전월세구분']].drop_duplicates()
```

3) index= 100~200, column = 전월세구분, 전용면적(㎡),보증금(만원)으로 이루어진 데이터프레임 추출
