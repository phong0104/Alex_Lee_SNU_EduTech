```python
import sklearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv('../data/Cars93.csv')
cars
```

```python
cars[['Length', 'Weight']]
```

```python
plt.scatter(x=cars['Length'], y=cars['Weight'], c='g', s=100, alpha=0.5)
plt.show()
```

```python
plt.title('Cars93', loc='right', pad = 10)
```

```python
title_font = {
    'fontsize':16,
    'fontweight':'bold'
}

plt.scatter(x=cars['Length'], y=cars['Weight'], alpha=0.5)
plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('Cars93', fontdict=title_font, loc='left', pad=20)
plt.show()
```

```python
a, b = cars['Length'].min(), cars['Length'].max()
print(a, b)
```

```python
c, d = cars['Weight'].min(), cars['Weight'].max()
print(c, d)
```

```python
plt.scatter(x=cars['Length'], y=cars['Weight'], alpha=0.5)
plt.axis([0,230,0,4400])
title_font = {
    'fontsize':20,
    'fontweight':'bold'
}
plt.xlabel('Length')
plt.ylabel('Weight')
plt.title('Cars93', fontdict = title_font, loc='left', pad=20)
plt.show()
```
