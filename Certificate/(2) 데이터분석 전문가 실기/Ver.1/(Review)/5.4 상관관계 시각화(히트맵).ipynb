```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air = pd.read_csv('../data/airquality.csv')
air
```

```python
air = air.iloc[:, :4]
air
```

```python
air_cor = air.corr(method='pearson') # default
air_cor
```

```python
air_cor2 = air.corr()
air_cor2
```

```python
air_cor.style.background_gradient(cmap='coolwarm')
```

```python
sns.heatmap(air_cor,
            xticklabels = air_cor.columns,
            yticklabels = air_cor.columns,
            cmap = 'RdBu_r',
            annot = True,
            linewidth = 3)
```
