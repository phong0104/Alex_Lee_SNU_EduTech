# 1.4 데이터 정렬

### 1.4.1 Index 기준 정렬
- df.sort_index() : df 객체 인덱스 기준으로 정렬
- default : 오름차순
- ascending = False : 내림차순

```python
import pandas as pd
```

```python
df_t = pd.read_csv('data/titanic.csv')
df_t.head()
```

```python
df_t.sort_index(axis=0, ascending=False)
```

```python
df_t.sort_index(ascending=False)
```

### 1.4.2 Column 기준 정렬
- df.sort_index(axis=1) : df 객체 column 명으로 정렬
- default : 오름차순
- ascending = False : 내림차순

```python
df_t.sort_index(axis=1, ascending = False)
```

### 1.4.3 값 기준 정렬
- obj.sort_values() : 시리즈 객체 값 기준으로 오름차순 정렬
- df.sort_values(by='컬럼명') : df 객체를 지정한 컬럼의 값 기준으로 정렬
- df.sort_values(by=['컬럼명1', '컬럼명2']) : df 객체를 지정한 컬럼들의 값 기준으로 정렬

```python
df_t.sort_values(by=['age', 'fare'], ascending=False).head(10)
```

```python
df_t.sort_values(by=['survived', 'age'], ascending=True)
# 오름차순이던, 내림차순이던 NaN은 맨 밑으로 간다
```
