### 3.1.1 특정 column을 기준으로 데이터를 그룹지어 집계함수 적용
- df.groupby('그룹화할 기준열').function()
- 2개 이상의 그룹화를 할 때에는 배열로 컬럼명 입력
- df.groupby(['컬럼명1', '컬럼명2']).function()

```python
import pandas as pd
df_iris = pd.read_csv('data/iris.csv')
df_iris
```

```python
df_iris.groupby(['target']) # 그룹화만 시키고 객체만 생긴다
```

```python
df_iris.groupby(['target']).sum()
# 그룹에만 보통 명목형 변수, 범주형 변수 (category)를 가지는 변수가 들어간다
```

```python
df_iris.groupby(['target']).mean()
# sepal : 꽃받침, petal : 꽃잎
```

```python
df_iris.target.unique()
```

```python
df_iris.groupby(['sepal length']).mean()
# 집계가 될 수 없는 변수(target)은 사라진다
```

```python
df_iris.groupby('target').mean()
```

```python
df_iris.groupby('target').mean().reset_index()
```

```python
# Q1. iris 데이터에서 종별 Speal.Width 평균 구하기
df_iris.groupby('target').mean()[['sepal width']] # [[]] : DataFrame으로 불러옴
```

```python
# Q2. iris 데이터에서 종별 Sepal.Width와 Petal.Width의 평균을 구해보자
avg2 = df_iris.groupby('target').mean()[['sepal width', 'petal width']]
avg2
```

#### 3.1.2.2 기준이 두개일 때
- pd.crosstab(Series1, Series2, dropna= , normalize= )
- dropna = True이면, 항목이 모두 NaN인 열은 포함하지 않음

```python
# Titanic에서 좌석등급과 생존여부의 관계를 살펴보기 위해 Class 변수에 따른 Survived 변수의 도수를 표 형태로 나타내시오
pd.DataFrame(df_t[['class', 'survived']].value_counts())
```

탐색적 데이터 분석(EDA)를 했다. 
- 별다른 기술을 쓰는 것이 아니라 데이터를 보고 인사이트를 얻을 수 있으면 된다.

```python
pd.crosstab(df_t['class'], df_t['survived'], dropna=True)
```

### 3.1.2.3 범주형 변수에 대한 상대도수 비율
- pd.crosstab(index, columns, dropna= , normalize= )
- normalize = True이면 모든 값에 대해 정규화
- index를 전달하면 행에 대해 정규화
- columns를 전달하면 열에 대해 정규화

Titanic 데이터의 sex 변수는 성별을 나타냄.

sex 변수에 따른 생존여부의 관계를 전체에 대한 비율, 행별 비율, 열별 비율로 살펴보기

```python
pd.crosstab(df_t['sex'], df_t['survived'])
```

```python
pd.crosstab(df_t['sex'], df_t['survived'], normalize=True)
```

```python
pd.crosstab(df_t['sex'], df_t['survived'], normalize=0)
```

```python
pd.crosstab(df_t['sex'], df_t['survived'], normalize='index') # 0과 같다
```

```python
pd.crosstab(df_t['sex'], df_t['survived'], normalize=1)
```

```python
pd.crosstab(df_t['sex'], df_t['survived'], normalize='columns') # 1과 같다
```
