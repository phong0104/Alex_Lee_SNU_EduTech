## 4.4 범주형 데이터 전처리하기
- 연속형 변수 : 수치형 변수
- 범주형 변수 : 이산형 or Category
- 범주형 변수를 표현하는 데 가장 널리 쓰이는 방법 : 원-핫 인코딩(one-hot-encoding)
- 범주형 변수에 있는 원소를 column으로 변경 >>> 해당 범주에 속하면 1, 아니면 0으로 채우는 기법
    - pd.get_dummies('범주형 변수')
    - pd.get_dummies(data.columns=['범주형1', '범주형2'])

```python
import pandas as pd
df = pd.read_csv('data/빅분기_X_train.csv', encoding='cp949')
df.head()
```

### 4.4.1 범주형변수를 인코딩하기

```python
df['주구매상품'].unique()
```

```python
len(df['주구매상품'].unique())
```

```python
len(pd.get_dummies(df['주구매상품']).columns)
```

```python
pd.get_dummies(df['주구매상품']).head()
```

### 4.4.2 원본 데이터에서 범주형 변수를 대체하기

```python
df_dum = pd.get_dummies(df, columns=['주구매상품', '주구매지점'])
df_dum.head()
```

```python
pd.concat([df, pd.get_dummies(df['주구매상품']), pd.get_dummies(df['주구매지점'])], axis=1).drop(columns=['주구매상품', '주구매지점']).head()
```
