# 3.2 데이터 모양 변경
### 3.2.1 데이터 재구조화(melt)
- pd.melt(df, id_vars=['A', 'D'], value_valrs=['B', 'E'], var_name='myVarName', value_name='myValName', ignore_index=False)

```python
import pandas as pd
import numpy as np

df_student = pd.DataFrame({'id':['S01', 'S02', 'S03', 'S04', 'S05'],
                           'kor':[80, 70, 56, 69, 90],
                           'math':[84,70,90,60,45],
                           'eng':[55,90,85,90,80]})
df_student
```

```python
df_student.info()
```

```python
melt_st = pd.melt(df_student, id_vars=['id'], value_vars=['kor','math','eng'],
                  var_name='subject', value_name='score', ignore_index=False)
melt_st
# id_vars 는 주로 object
# value_vars 는 주로 집계가능한 함수
```

### 3.2.2 데이터 재구조화(pivot_table)
- pd.pivot_table(data, index, columns, values, aggfunc, margins)

```python
pd.pivot_table(melt_st, index='id', columns='subject', values='score').head(10)
```

```python
df_fruits = pd.read_csv('data/Fruits.csv')
df_fruits
```

```python
melt_fruits = pd.melt(df_fruits, id_vars = ['Fruit'], value_vars = ['Sales', 'Expenses', 'Profit'])
melt_fruits
```

```python
melt_fruits.pivot_table(index=['Fruit'], columns='variable', 
                        values='value', aggfunc='mean', margins=False)
# groupby로도 만들수는 있지만, pivot table이 만능이다
```

```python
melt_fruits.pivot_table(index=['Fruit'], columns='variable', 
                        values='value', aggfunc='sum', margins=True)
# margin 이 True이면 전체 합계를 나타내준다
```

### 3.2.3 데이터 행, 열 변경(transpose)

```python
melt_fruits.pivot_table(index=['variable'], columns='Fruit', 
                        values='value', aggfunc='sum', margins=True)
# 이렇게 해줘도 되지만...
```
