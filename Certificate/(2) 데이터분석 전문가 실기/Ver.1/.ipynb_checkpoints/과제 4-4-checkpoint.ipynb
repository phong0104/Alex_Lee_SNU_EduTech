```python
## 1. 타이타닉 데이터의 결측치를 확인하세요
import pandas as pd
df_titanic = pd.read_csv('data/titanic.csv')
df_titanic.info()
```

```python
## 2. age의 결측값은 중위 값으로 대체하세요. 
print(df_titanic['age'].median())
df_titanic['age'].fillna(df_titanic['age'].median(), inplace=True)
df_titanic.info()
```

```python
df_titanic.head()
```

```python
## 3. embark_town의 결측 값은 최빈값으로 대체하세요.
df_titanic['embark_town'].value_counts()
```

- 방법1

```python
df_titanic['embark_town'].fillna('Southampton', inplace= True)
```

- 방법2

```python
print(df_titanic['embark_town'].mode()[0])
df_titanic['embark_town'].fillna(df_titanic['embark_town'].mode()[0], inplace=True)
```

```python
df_titanic.info()
```

```python
## 4. embarked 결측치는 embark_town의 첫 글자로 대체하세요
print(df_titanic['embark_town'].str[0])
df_titanic['embarked'].fillna(df_titanic['embark_town'].str[0], inplace=True)
df_titanic.head()
```

```python
df_titanic.info()
```

```python
## 5. df_t내에 범주형 변수(embark_town,class) 를 원 핫 인코딩하여 더미변수로 만드세요
df2 = pd.get_dummies(df_titanic, columns=['embark_town', 'class'])
df2.head()
```
