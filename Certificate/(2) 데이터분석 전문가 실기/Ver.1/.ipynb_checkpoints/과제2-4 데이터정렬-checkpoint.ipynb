## players_20.csv데이터에서 height_cm, weight_kg 의 중위값(median)을 출력하세요.

```python
import pandas as pd 
```

```python
fifa_df = pd.read_csv("data/players_20.csv")
```

```python
fifa_df.head()
```

```python
fifa_df.columns
```

```python
height_medi = fifa_df['height_cm'].median()
height_medi
```

```python
weight_medi = fifa_df['weight_kg'].median()
weight_medi
```

## 데이터셋의 height_cm, weight_kg 이 중위값(median)보다 큰 선수를 출력하세요 

```python
fifa_df[ (fifa_df['height_cm'] > height_medi) & (fifa_df['weight_kg'] > weight_medi) ]
```

## height_cm 기준으로 내림차순 정렬하세요

```python
fifa_df.sort_values(by='height_cm', ascending=False)
```

## weight_kg 기준으로 내림차순 정렬하세요

```python
fifa_df.sort_values(by='weight_kg', ascending=False)
```

## height_cm 와 weight_kg 를 기준으로 오름차순 정렬하세요 

```python
fifa_df.sort_values(by=['height_cm', 'weight_kg'])
```
