2) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경

```python
data.loc[data['시군구'] == '세종특별자치시 고운동', 'new_시군구'] = '세종특별자치시 해밀동'
```

```python
data.head()
```

3) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용)

```python
data.loc[data['전월세구분'] == '전세', 'Class'] = True
data.loc[data['전월세구분'] != '전세', 'Class'] = False
data.head()

# data['Class'] = np.where(data['전월세구분'] == '전세', True, False)
```

4) 전세이면서 보증금이 1억 이상인 데이터만을 추출

```python
data.loc[(data['전월세구분'] == '전세') & (data['보증금(만원)'] >= 10000) ].head()
```

5) - 10 ~ 0층 : 지하 , 1 ~ 5층 : 저층 6~15층 : 중층, 16층 이상 : 고층으로 할당하는 새로운 컬럼 생성 

```python
import numpy as np

condition = [
    ( (data['층'] >= -10) & (data['층'] <= 0) ),
    ( (data['층'] >= 1) & (data['층'] <= 5) ),
    ( (data['층'] >= 6) & (data['층'] <= 15) ),
    ( data['층'] >= 16 )
]

choice = ['지하', '저층', '중층', '고층']

data['층구분'] = np.select(condition, choice, default = 'Not_Specified')
data.head(10)
```
