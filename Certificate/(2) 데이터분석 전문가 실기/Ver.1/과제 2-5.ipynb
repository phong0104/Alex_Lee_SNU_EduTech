##  players_19.csv와 players_20.csv 데이터를 행결합해서 master data를 만드세요 

```python
import pandas as pd 
```

```python
fifa_df19 = pd.read_csv("data/players_19.csv")
fifa_df20 = pd.read_csv("data/players_20.csv")
```

## fifa_df19 = players_19.csv에서 sofifa_id','short_name','overall' 컬럼만 추출
## fifa_df20 = players_20.csv 에서 sofifa_id와 overall만을 추출 
## sofifa_id를 key값으로  위의 두 테이블을 innerjoin 하세요 

## sofifa_id를 key값으로  위의 두 테이블을 outerjoin 하세요 

## outerjoin을 한 테이블 중 inner join을 한 테이블과 중복되는 행을 삭제하세요 

```python
fifa_df19.head()
```

```python
fifa_df20.head()
```

### 1) fifa_df19 = players_19.csv에서 sofifa_id','short_name','overall' 컬럼만 추출

```python
fifa_df19 = fifa_df19[ ['sofifa_id', 'short_name', 'overall'] ]
fifa_df19.head()
```

### 2) fifa_df20 = players_20.csv 에서 sofifa_id와 overall만을 추출

```python
fifa_df20 = fifa_df20[ ['sofifa_id', 'overall'] ]
fifa_df20.head()
```

### 3) sofifa_id를 key값으로 위의 두 테이블을 innerjoin 하세요

```python
merge_inner = pd.merge(fifa_df19, fifa_df20, how='inner', on='sofifa_id')
merge_inner
```

### 4) sofifa_id를 key값으로 위의 두 테이블을 outerjoin 하세요

```python
merge_outer = pd.merge(fifa_df19, fifa_df20, how='outer', on='sofifa_id')
merge_outer
```

### 5) outerjoin을 한 테이블 중 inner join을 한 테이블과 중복되는 행을 삭제하세요

- 방법1

```python
inner_series = merge_inner['sofifa_id']
```

```python
outer_series = merge_outer['sofifa_id']
```

```python
outer_series.isin(inner_series)
```

```python
merge_outer[ ~(outer_series.isin(inner_series)) ]
```

- 방법2

```python
pd.concat([merge_inner, merge_outer]).drop_duplicates(keep=False)
```

- 방법3

```python
idx = merge_inner['sofifa_id'].values
idx
```

```python
merge_outer.set_index('sofifa_id', inplace=True)
merge_outer
```
