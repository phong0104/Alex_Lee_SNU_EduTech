## 전세의 보증금 평균 값을 구하세요 groupby 사용 


```python
import pandas as pd
df = pd.read_csv("data/7.세종시_아파트(전월세)_실거래가.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.groupby('전월세구분').mean()['보증금(만원)']
```

```python
print(df.groupby('전월세구분').mean()['보증금(만원)'][1])
```

## 계약년도별, 시군구별 평균 보증금 가격를 구하세요  



```python
df_mean = df.groupby(['계약년월', '시군구'])['보증금(만원)'].mean()
pd.DataFrame(df_mean)
```

```python
def get_year(df):
    return df // 100

df['계약년도'] = df['계약년월'].apply(get_year)
df.head()
```

```python
# df["계약년월"]=df["계약년월"].astype(str)
# df["계약년"]=df["계약년월"].str[:4]
# df
```

```python
pd.DataFrame(df.groupby(['계약년도', '시군구'])['보증금(만원)'].mean())
```

## 계약년도별 전세 건수를 구하세요 

- 방법1

```python
pd.crosstab(df['계약년도'], df['전월세구분'], dropna=True)
```

```python
pd.crosstab(df['계약년도'], df['전월세구분'], dropna=True)['전세'].values
```

- 방법2

```python
df.groupby('계약년도')['전월세구분'].value_counts()
```

```python
df.groupby('계약년도')['전월세구분'].value_counts().values[0::2]
```
