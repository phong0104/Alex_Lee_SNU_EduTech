# 날짜 데이터 전처리 과제 
- 데이터 : "6.세종시_오피스텔(매매)_실거래가.csv"

```python
import pandas as pd 
df_sejong = pd.read_csv("data/6.세종시_오피스텔(매매)_실거래가.csv",encoding='utf8')
```

```python
from datetime import datetime
datetime.strptime('20220111', '%Y%m%d')
```

```python
df_sejong.head()
```

```python
### 아래와 같이 계약년월을 datetime으로 변경하세요. 
```

```python
df_sejong.info()
```

```python
df_sejong['계약년월'] = df_sejong['계약년월'].astype(str)
df_sejong['계약일'] = df_sejong['계약일'].astype(str)
df_sejong.info()
```

```python
df_sejong['계약년월'] = df_sejong['계약년월'] + df_sejong['계약일']
df_sejong.head()
```

```python
df_sejong = df_sejong.drop(['계약일'], axis=1)
df_sejong.head()
```

```python
df_sejong.rename(columns={'계약년월':'계약날짜'}, inplace=True)
df_sejong.head()
```

```python
df_sejong['계약날짜'] = pd.to_datetime(df_sejong['계약날짜'], format='%Y%m%d')
df_sejong.head()
```

```python
df_sejong.info()
```

```python
## 계약년월을 아래와 같이 문자열로 다시 변경하세요. 
```

```python
df_sejong['계약날짜'] = df_sejong['계약날짜'].dt.strftime('%Y_%m_%d')
df_sejong.head()
```

```python
df_sejong.info()
```
