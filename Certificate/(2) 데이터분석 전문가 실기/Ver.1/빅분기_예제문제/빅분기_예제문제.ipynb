# 작업형 1유형 예시문제 

* mtcars 데이터셋의 qsec 컬럼을 최소최대 척도(Min Max Scale)로 변환한 후 0.5 보다 큰 값을 가지는 레코드 수를 구하시오.

```python
import pandas as pd
cars = pd.read_csv('mtcars.csv', encoding='cp949')
cars.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(cars[['qsec']]))
df.head()
```

```python
(df>0.5).head()
```

```python
np.sum(df>0.5)
```

# 2. 작업형 제2유형 : 모형 구축 및 평가 영역
- 고객 3,500명에 대한 학습용 데이터(y_train.csv, X_train.csv)를 이용하여 성별예측 모형을 만든 후, 이를 평가용 데이터(X_test.csv)에 적용하여 얻은 2,482명 고객의 성별 예측값(남자일 확률 = gender:1)을 다음과 같은 형식의 CSV 파일로 생성하시오.(제출한 모델의 성능은 ROC-AUC 평가지표에 따라 채점)



```python
X_train = pd.read_csv('X_train.csv',encoding='cp949')
y_train = pd.read_csv('y_train.csv',encoding='cp949')
X_test = pd.read_csv('X_test.csv',encoding='cp949')

X_train.head()
```

```python
X_train.info()
```

```python
X_test.info()
```

```python
X_train.shape
```

```python
X_train.set_index('cust_id', inplace=True)
X_train.head()
```

```python
X_train = pd.get_dummies(X_train, columns = ['주구매상품', '주구매지점'])
X_train.head()
```

```python
y_train.set_index('cust_id', inplace=True)
y_train.head()
```

```python
y_train.shape
```

```python
X_test.set_index('cust_id', inplace=True)
X_test.head()
```

```python
X_test = pd.get_dummies(X_test, columns=['주구매상품', '주구매지점'])
```

```python
X_test.head()
```

```python
X_test['주구매상품_소형가전'] = 0
X_test.shape
```

```python
X_train.head()
```

```python
np.sum(X_train['환불금액']==0)
```

```python
X_train['환불금액'] = X_train['환불금액'].fillna(0)
np.sum(X_train['환불금액'].isna())
```

```python
X_test['환불금액'] = X_test['환불금액'].fillna(0)
np.sum(X_test['환불금액'].isna())
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train['gender'])
```

```python
clf.predict_proba(X_test)
```

```python
clf.predict_proba(X_test)[:,1]
```

```python
result = clf.predict_proba(X_test)[:,1]
pd.DataFrame(result)
```

```python
pd.DataFrame(result).to_csv('result.csv', index=False)
```
