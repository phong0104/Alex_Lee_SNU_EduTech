## 조건에 따른 row/column 선택 및 삭제 
### 조건에 따른 column 선택  

    조건이 하나일 때 : df[df['컬럼명'] >= 40]
    조건이 두개 이상 : df[(조건1) & (조건2) & ((조건3) | (조건4)) ]
    ** 관계연산자의 우선순위를 모른다면 그냥 괄호를 많이 쓸 것 
  

```python
import pandas as pd 
import numpy as np

df_t = pd.read_csv("titanic.csv")
df_t.drop(columns=['alive','deck','embark_town'],inplace=True)
df_t.head()
```

```python
## 하나의 조건문 
df_t[df_t["survived"]==1].head()
```

```python
## 조건이 여러개 
df_t[ (df_t["survived"]==1) | (df_t['fare']>30) & (df_t['class']=="First") ]
```

### 조건에 따른 column 추가, 변경

```python
df_t.loc[df_t["survived"]==1,'temp'] =99999
df_t
```

```python
df_t["temp2"] = [True if i > 30 else False for i in df_t['age']]
```

```python
df_t['temp3']=False
df_t.loc[df_t.age>30,'temp3']
```

```python
df_t.loc[df_t.age>30,'temp3'] = 50
```

```python
df_t
```

```python
## 데이터 프레임으로 접근하여 값을 대입하면 오류가 남 // 데이터 프레임 접근은 복사본이므로 데이터에 직접 접근할 때에는 loc를 써야함 
df_t[df_t["age"]>30].temp3 = True
```

```python
df_t[df_t["age"]>30].temp3
```

```python
df_t.drop(columns=['temp','temp2','temp3'],inplace=True)
```

#### np.where()
* 데이터 프레임 or 리스트 내 필드에 접근하고 수정하는 방법 . 
* np.where(condition, x, y)는 조건이 충족되면x를, 그렇지 않으면y를 반환합니다.

```python
df_t['alive'] = np.where(df_t['survived']==1, 'Yes', 'No')
df_t.head(10)
```

#### np.select()
* 데이터 프레임 or 리스트 내 필드에 접근하고 수정하는 방법 . 
* 조건 목록과 선택 목록을 입력으로 받아서 조건에 따라 선택 목록의 요소로 구성된 배열을 반환합니다. 
* 이 방법을 사용하면 두 개 이상의 조건이있을 때 Pandas의 주어진 조건을 기반으로 DataFrame 열을 만들 수 있습니다.

```python
conditionlist = [
    (df_t['age'] >= 19) ,
    (df_t['age'] >= 8) & (df_t['age'] <19),
    df_t['age'] < 8]
choicelist = ['성인', '초중고등학생', '미취학아동']

df_t['status'] = np.select(conditionlist, choicelist,default='Not_Specified')

df_t.head(10)
```

### 데이터 정렬
#### Index 기준 정렬
    df.sort_index() : df 객체 인덱스 기준으로 정렬 
    df.sort_index(axis=1) : df 객체 column 명으로 정렬 

```python
df_t.sort_index(axis=1)
```

#### 값 기준 정렬 
    obj.sort_values() :시리즈 객체 값 기준으로 오름차순 정렬 
    df.sort_values(by='컬럼명') : df객체를 지정한 컬럼의 값 기준으로 정렬 
    df.sort_values(by=['컬럼명1','컬럼명2]) : df객체를 지정한 컬럼들의 값 기준으로 정렬
    
    default : 오름차순 
    ascending = False : 내림차순 

```python
df_t.sort_values(by='age', ascending=False).head(10)
```
