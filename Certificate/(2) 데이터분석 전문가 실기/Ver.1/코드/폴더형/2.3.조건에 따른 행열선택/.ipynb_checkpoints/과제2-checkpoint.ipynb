# 과제 2주차  
    데이터 : 세종시_아파트(전월세)_실거래가.csv
    1) 세종시 데이터에서 컬럼 확인 후, unnamed 컬럼 제거
    2) 전월세구분 컬럼에서 unique값 확인  
    3) 전세만으로 이루어진 데이터프레임 추출 
    4) 단지명 중  '해밀마을 2단지'인 row 삭제 
    5) new_시군구' 컬럼을 만들고 '세종특별자치시 고운동' => '세종특별자치시 해밀동' 으로 변경
    6) 'Class'컬럼을 만든 뒤 전세면 True 아니면 False로 입력 (loc사용) 

```python
import pandas as pd 
```

```python
df_sejong = pd.read_csv('세종시_아파트(전월세)_실거래가.csv')
```

```python
df_sejong['new_시군구'] = df_sejong['시군구']
```

```python
df_sejong
```

```python
df_sejong.loc[df_sejong["시군구"] =='세종특별자치시 해밀동',['new_시군구']] = '세종특별자치시 고운동'
```

```python
df_sejong.loc[df_sejong["시군구"] =='세종특별자치시 해밀동','new_시군구'] = "해밀동입니다"
```
