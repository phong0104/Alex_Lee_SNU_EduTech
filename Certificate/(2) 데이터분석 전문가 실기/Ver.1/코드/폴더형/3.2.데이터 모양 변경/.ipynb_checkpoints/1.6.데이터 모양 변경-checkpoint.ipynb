# 데이터 모양 변경

## 데이터 재구조화(melt)
    pd.melt(data, id_vars, var_name, value_name) 

```python
import pandas as pd
import numpy as np
data = pd.read_csv('credit.csv')

data
```

```python
data.info()
```

```python
data_melted = pd.melt(data, ['Age (years)','Purpose'], var_name= 'columns', value_name='value')
data_melted[1001::]
```

```python
data_melted.pivot_table(index=['Age (years)'], columns=['columns'], values='value' , aggfunc = np.mean ).tail(20)
```

```python
data
```

```python
data.pivot_table(index=['Age (years)',"Instalment per cent"],columns=["Creditability"],values="Credit Amount",aggfunc=np.mean,margins=True)
```

## 데이터 재구조화(pivot_table)
    pd.pivot(data, index, columns, values, aggfunc)
    pd.pivot_table(data, index, columns, values, aggfunc, margins)
    # index가 2개 이상일 경우 pivit_table은 작동하지만 data.pivot 은 작동 안함!
    # columns 가 2개 이상인 경우 pivot_table은 작동하지만 data.pivot 은 작동 안함!
    * 결론 : pivot_table만 익힐 것 

```python
pd.pivot_table(data, index='Age (years)', columns='Purpose', values='Credit Amount').head(10)
```

```python
# index가 2개 이상일 경우 pivit_table은 작동하지만 data.pivot 은 작동 안함!
data.pivot_table(index=['Age (years)', 'Foreign Worker'], columns='Purpose', values='Credit Amount')
```

```python
 pd.pivot_table(data, index='Age (years)', columns=['Foreign Worker', 'Purpose'], values='Credit Amount').head()
```

```python
# columns 가 2개 이상인 경우 pivot_table은 작동하지만 data.pivot 은 작동 안함!
x = pd.pivot_table(data, index='Age (years)', columns=['Foreign Worker', 'Purpose'], values='Credit Amount')
```

```python
## 멀티인덱스로 컬럼이 구성되어 있음 
x.columns
```

```python
x[(1,  0)].head()
```

```python
# index 중복값이 있으면 pivot_table은 작동하지만 data.pivot은 작동 안함!
data.pivot_table(index='Age (years)', columns=['Foreign Worker', 'Purpose'], values='Credit Amount',aggfunc=np.sum).head(4)
```

```python
# pd.pivot_table 은 margins=True로 설정하면 행과 열을 기준으로 합계를 같이 제시해줌
pd.pivot_table(data, index='Age (years)', columns='Purpose',
               values='Credit Amount', aggfunc=np.mean, margins=True).tail(10)
```
