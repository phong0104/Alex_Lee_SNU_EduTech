# 데이터 모양 변경

## 데이터 재구조화(melt)
    pd.melt(data, id_vars, var_name, value_name) 

```python
import pandas as pd
import numpy as np

df_student = pd.DataFrame({'id':['S01', 'S02', 'S03', 'S04','S05'],
                        'kor':[80,75,56,69,90],
                         'math' :[84,70,90,60,45],
                         'eng' : [55,90,85,90,80]})

df_student
```

```python
df_student.info()
```

```python
melt_st= pd.melt(df_student,['id'])
melt_st
```

## 데이터 재구조화(pivot_table)
    pd.pivot_table(data, index, columns, values, aggfunc, margins)

```python
pd.pivot_table(melt_st, index='id', columns='variable', values='value').head(10)
```

```python
df_fruits = pd.read_csv('../data/Fruits.csv')
df_fruits
```

```python
melt_fruits = pd.melt(df_fruits,['Fruit'],value_vars=['Sales','Expenses'])
melt_fruits
```

```python
melt_fruits.pivot_table(index=['Fruit'], columns='variable', values='value', aggfunc= 'mean', margins=False)
```

```python
melt_fruits.pivot_table(index=['Fruit'], columns='variable', values='value', aggfunc='sum', margins=True)
```

## Transpose
* 데이터의 행, 열 변경 

```python
df_fruits
```

```python
df_fruits.transpose()
```

```python
melt_fruits.pivot_table(index=['Fruit'], columns='variable', values='value', aggfunc='sum', margins=True).transpose()
```
