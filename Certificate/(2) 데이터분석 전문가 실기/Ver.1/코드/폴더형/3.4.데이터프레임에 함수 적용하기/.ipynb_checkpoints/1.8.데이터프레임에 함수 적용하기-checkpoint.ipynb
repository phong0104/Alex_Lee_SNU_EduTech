## 데이터프레임에 함수 적용하기 
### 행 또는 열 방향으로 주어진 함수를 한 번에 적용 : apply 


```python
import pandas as pd
import numpy as np
```

```python
df_sejong = pd.read_csv('세종시_아파트(전월세)_실거래가.csv',encoding='utf8')
df_sejong 
```

```python
df_sejong.drop(columns='Unnamed: 0',inplace = True )
df_sejong
```

```python
df_sejong
```

```python
for i in range(len(df_sejong)) : 
    if df_sejong['전용면적(㎡)'][i] > df_sejong['보증금(만원)'][i] : 
        temp = df_sejong['전용면적(㎡)'][i]
    else : 
        temp = df_sejong['보증금(만원)'][i]
    df_sejong['new_columns'][i] = temp
```

```python
df_sejong[['전용면적(㎡)','보증금(만원)']]
```

```python
## 열 방향일때는 그냥 집계함수를 쓰면 되지만 행방향 비교일 때는 apply가 훨씬 편하다 
## 시험장에서 기억이 안나면 그냥 노가다 해라 (빅분기)
## apply 레퍼런스 참고 

df_sejong[['전용면적(㎡)','보증금(만원)']].apply(np.max,axis= 1)
```

### 데이터 내부 각 요소를 처리하는 함수 만들기 
    data['col_name'] = data['col_name'].map(lambda x:x.func(), na_action = 'ignore') 
    map(lambda)를 사용할 때에는 시리즈 형식이어야 함 . 
    map은 원소 접근이어햐 함 (시리즈)  1:1 매핑  대체 원소 -> 원소.func()  
    lambda는 간단한 함수 개념 

```python
df_sejong['시군구'].str.split(" ",expand=True)[1]
```

```python
df_sejong['시군구'].map(lambda x:x.split(" ")[1]) 
```

```python
temp_list = [1,2,3,4,5]
list(map(lambda x:x*10,temp_list))
```
