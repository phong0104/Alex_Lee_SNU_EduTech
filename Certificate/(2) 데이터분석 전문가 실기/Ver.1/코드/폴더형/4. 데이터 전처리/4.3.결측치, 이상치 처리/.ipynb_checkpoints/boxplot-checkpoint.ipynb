### 3절. 이상치 인식
#### 1. 사분위수
* 데이터명.describe()
    * 결측치가 아닌 데이터행의 개수
    * 평균/표준편차
    * 최소/최댓값
    * Q1, Q2, Q3 나타내줌
* IQR을 이용한 이상치 탐색 : 함수 생성

```python
import pandas as pd 
import numpy as np 
```

```python
air = pd.read_csv('../data/airquality.csv')
print(air.head())
```

```python
air
```

```python
# Ozone 변수에서 이상치를 박스플롯으로 확인
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='Ozone', data=air)
plt.show() 
```

```python
air.quantile(0.25)
```

```python
# IQR 을 이용한 이상치 탐색 함수

def detect_outliers(df=None, column=None, weight=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    IQR = Q3 - Q1
    IQR_weight = IQR*weight
    
    outlier_idx = df[(df[column] < Q1 - IQR_weight) | (df[column] > Q3 + IQR_weight)].index
    
    return outlier_idx
```

```python
out_index = detect_outliers(df=air, column='Ozone')
```

```python
air.loc[out_index,'outlier']=True
```

```python
air[air['outlier']==True]
```

```python
Outliers_to_drop = detect_outliers(df=air, column='Ozone', weight=1.5)
```

```python
# IQR로 이상치를 찾아 인덱스 저장
Outliers_to_drop
```

```python
# Ozone 변수에서 outliers 가 있었던 데이터 확인
air.loc[Outliers_to_drop]
```

```python
# outlier 제거한 데이터
air_2 = air.drop(Outliers_to_drop, axis=0).reset_index(drop=True)
```

```python
air_2
```

```python
sns.boxplot(x='Ozone', data=air_2)
plt.show() 
```
