```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt
```

```python
US = pd.read_csv('../USArrests.csv')
US.head()
```

```python
US.columns = ['State', 'Murder', 'Assault', 'UrbanPop', 'Rape']
labelList = US.State.tolist()
US.head()
```

```python
labelList
```

```python
single = linkage(US.iloc[:, 1::], metric = 'euclidean', method = 'single')
```

```python
single
```

```python
plt.figure(figsize=(10,7))
dendrogram(single,
           orientation = 'top',
           labels = labelList,
           distance_sort = 'descending',
           color_threshold = 25,
           show_leaf_counts = True)
plt.axhline(y=25, color = 'r', linewidth = 1)
plt.show()
```

```python
plt.figure(figsize=(10,7))
dendrogram(single,
           orientation = 'top',
           labels = labelList,
           distance_sort = 'ascending',
           color_threshold = 25,
           show_leaf_counts = True)
plt.axhline(y=25, color = 'r', linewidth = 1)
plt.show()
```

```python
plt.figure(figsize=(10,7))
dendrogram(single,
           orientation = 'bottom',
           labels = labelList,
           distance_sort = 'ascending',
           color_threshold = 25,
           show_leaf_counts = True)
plt.axhline(y=25, color = 'r', linewidth = 1)
plt.show()
```

```python
assignments = fcluster(single, 25, 'distance')
assignments
```

```python
US['cluster'] = assignments
```

```python
US.head()
```

```python
complete = linkage(US.iloc[:, 1::], metric = 'euclidean', method = 'complete')

plt.figure(figsize = (10, 7))
dendrogram(complete,
           orientation = 'top',
           labels = labelList,
           distance_sort = 'descending',
           color_threshold = 100,
           show_leaf_counts = True)
plt.axhline(y=100, color = 'r', linewidth = 1)
plt.show()
```

```python
ward = linkage(US.iloc[:, 1::], metric = 'euclidean', method = 'ward')

plt.figure(figsize=(10,7))
dendrogram(ward,
           orientation = 'top',
           labels = labelList,
           distance_sort = 'descending',
           color_threshold = 250,
           show_leaf_counts = True)
plt.axhline(y=250, color='r', linewidth = 1)
plt.show()
```

```python
ward = linkage(US.iloc[:, 1::], metric = 'euclidean', method = 'ward')

plt.figure(figsize=(10,7))
dendrogram(ward,
           orientation = 'top',
           labels = labelList,
           distance_sort = 'descending',
           color_threshold = 250,
           show_leaf_counts = False)
plt.axhline(y=250, color='r', linewidth = 1)
plt.show()
```

```python
assignments = fcluster(ward, 250, 'distance')
assignments
```

```python
US['cluster'] = assignments
```

```python
US.groupby('cluster').mean()
```
